```python
file = open('data/Advertising.csv')
```

```python
# leemos el contenido del archivo csv
contents = csv.reader(file)

# query de SQL para insertar los valores de la tabla dentro de la tabla
insert_records = "INSERT INTO datos VALUES(?, ?, ?, ?)"

cursor.executemany(insert_records, contents)
```

```python
# Query de SQL para verificar los datos insertados en la tabla

borrar_encabezado_extra = "DELETE FROM datos WHERE TV = 'TV'"
rows_2 = cursor.execute(borrar_encabezado_extra).fetchall()
  
# Output to the console screen
for r in rows_2:
    print(r)
  
# Commiting the changes
connection.commit()
```

```python
# Query de SQL para verificar los datos insertados en la tabla
select_all = "SELECT * FROM datos"
rows = cursor.execute(select_all).fetchall()
  
# Output to the console screen
for r in rows:
    print(r)
  
# Commiting the changes
connection.commit()
```

```python
def sql_query(query):

    # Ejecuta la query
    cursor.execute(query)

    # Almacena los datos de la query 
    ans = cursor.fetchall()

    # Obtenemos los nombres de las columnas de la tabla
    names = [description[0] for description in cursor.description]

    return pd.DataFrame(ans,columns=names)

query = '''
SELECT * 
FROM datos
'''


df = sql_query(query)
df
```
