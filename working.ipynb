```python
import csv
import sqlite3
import pandas as pd
```

```python
# Creamos la base de datos
connection = sqlite3.connect('advertising.db')

# Creamos un cursor
cursor = connection.cursor()
```

```python
create_table = '''CREATE TABLE datos(
                TV INTEGER NOT NULL,
                radio INTEGER NOT NULL,
                newspaper INTEGER NOT NULL,
                sales INTEGER NOT NULL);
                '''


# Creamos la tabla dentro de nuestra base de datos
cursor.execute(create_table)
```

```python
file = open('data/Advertising.csv')
```

```python
# leemos el contenido del archivo csv
contents = csv.reader(file)

# query de SQL para insertar los valores de la tabla dentro de la tabla
insert_records = "INSERT INTO datos (TV, radio, newspaper, sales) VALUES(?, ?, ?, ?)"

cursor.executemany(insert_records, contents)
```

```python
# Query de SQL para verificar los datos insertados en la tabla
select_all = "SELECT * FROM datos"
rows = cursor.execute(select_all).fetchall()
  
# Output to the console screen
for r in rows:
    print(r)
  
# Commiting the changes
connection.commit()
```
